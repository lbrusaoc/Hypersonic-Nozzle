```python
import bezier
import numpy as np
import matplotlib.pyplot as plt

# Define points
node1 = np.array([0.0, 0.25])
ctrl1 = np.array([0.25, node1[1]])   # (0.25, 0.25)

node2 = np.array([0.5, 0.15])
ctrl2 = np.array([0.25, node2[1]])   # (0.25, 0.15)

# For a quadratic bezier curve we need 3 control points.
# We can interpolate through node1 -> ctrl1 -> node2
nodes = np.column_stack([node1, ctrl1, node2])

curve = bezier.Curve(nodes, degree=2)

# Plot
s_vals = np.linspace(0.0, 1.0, 200)
curve_points = curve.evaluate_multi(s_vals)

plt.plot(curve_points[0, :], curve_points[1, :], label="Bezier curve")
plt.scatter(nodes[0, :], nodes[1, :], color='red', label='Control points')
plt.legend()
plt.axis("equal")
plt.show()
```
